```python
def data_info(data):
  cols = []
  d_types = []
  n_unique = []
  unique_val = []
  null = []
  for col in data.columns:
    cols.append(col)
    d_types.append(data[col].dtypes)
    unique_val.append(data[col].unique())
    n_unique.append(data[col].nunique())
    null.append(data[col].isnull().sum())
  return pd.DataFrame({'columns':cols,'Unique Value':unique_val,"no_Unique":n_unique,'Data Types':d_types ,'Nulls':null})


```

```python
data_info(flights_data)
```

# Data Preprocessing

```python
flights_data.dropna(subset=['TAIL_NUMBER','SCHEDULED_TIME'],inplace=True)

flights_data['CANCELLATION_REASON'].fillna('Not Cancelled', inplace=True)

delay_col = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']
flights_data[delay_col] = flights_data[delay_col].fillna(0)
```

```python
data_info(flights_data)
```

```python
origin_airport_encoder = pd.merge(origin_airport_A,origin_airport_N, left_on='Standardized_Airport', right_on='Standardized_Airport_N', how='inner')
origin_airport_encoder.drop(columns=['Description_x', 'Standardized_Airport', 'Description_y', 'Standardized_Airport_N'], inplace=True)

origin_airport_encoder
```

```python
flights_data = flights_data.merge(origin_airport_encoder[['Airport_Code', 'Airport_Code_N']], how='left', left_on='ORIGIN_AIRPORT', right_on='Airport_Code')
```

```python
flights_data
```

```python
flights_data = flights_data.drop_duplicates(subset=['MONTH'	,'DAY','AIRLINE', 'FLIGHT_NUMBER', 'TAIL_NUMBER', 'ORIGIN_AIRPORT',	'DESTINATION_AIRPORT',	'SCHEDULED_TIME',	'Flight_Status','CANCELLATION_REASON', 'Airport_Code'], keep='last')
```

```python
flights_data
```

```python
flights_data.drop(columns=['ORIGIN_AIRPORT', 'Airport_Code'], inplace=True)
flights_data.rename(columns={'Airport_Code_N': 'ORIGIN_AIRPORT'}, inplace=True)

flights_data
```

```python
flights_data['ORIGIN_AIRPORT'].fillna(flights_data['ORIGIN_AIRPORT'].mode()[0], inplace=True)  
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
flights_data['CANCELLATION_REASON'] = le.fit_transform(flights_data['CANCELLATION_REASON'])

```

```python
flights_data_encoded = pd.get_dummies(flights_data, columns=['AIRLINE'], drop_first=True)
```

```python
boolean_columns = flights_data_encoded.select_dtypes(include=['bool']).columns
flights_data_encoded[boolean_columns] = flights_data_encoded[boolean_columns].astype(int)
```

```python
import category_encoders as ce

target_encoder = ce.TargetEncoder()
flights_data_encoded['TAIL_NUMBER'] = target_encoder.fit_transform(flights_data_encoded['TAIL_NUMBER'], flights_data_encoded['Flight_Status'])
flights_data_encoded['FLIGHT_NUMBER'] = target_encoder.fit_transform(flights_data_encoded['FLIGHT_NUMBER'], flights_data_encoded['Flight_Status'])
flights_data_encoded['DESTINATION_AIRPORT'] = target_encoder.fit_transform(flights_data_encoded['DESTINATION_AIRPORT'], flights_data_encoded['Flight_Status'])


```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)

flights_data_resampled = pd.concat([x_resampled, y_resampled], axis=1)

print(flights_data_resampled['Flight_Status'].value_counts(normalize=True))
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report, confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV

x=np.arange(1,10)
y=np.arange(10,100,10)

parameters = {'max_depth':(x), 'n_estimators':(y), 'min_samples_split':(y), 'min_samples_leaf':(x), 'max_leaf_nodes':(y)}

rf=RandomForestClassifier()
GR= GridSearchCV(rf, parameters)
GR.fit(x_train,y_train)

print(GR.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,
    min_samples_split=10,
    min_samples_leaf=4,
    max_depth=10,
    max_leaf_nodes=50,
    random_state=42)

model.fit(x_train, y_train)
```

# Evaluate the model

```python
y_pred = model.predict(x_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
accuracy 
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
print("CV Accuracy: ", np.mean(cv_scores))
```

```python
class_report = classification_report(y_test, y_pred,output_dict=True)
report = pd.DataFrame(class_report).transpose()
report
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')


plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
# L1 Regularization (Lasso)
model_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=1.0)  
model_l1.fit(x_train, y_train)

y_pred_l1 = model_l1.predict(x_test)

print("L1 Regularization Accuracy:", accuracy_score(y_test, y_pred_l1))
```
